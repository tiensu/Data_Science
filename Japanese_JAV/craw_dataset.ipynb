```python
import os
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Craw data

```python
ja_names = []
for page in np.arange(1,52):
    # get html table code
    base_page_url = 'https://japanpornjav.com/jav/index.html?page='
    req_resp_page = requests.get(base_page_url + str(page)).text
    soup_page_obj = BeautifulSoup(req_resp_page,'lxml')
    ja_names_h6 = soup_page_obj.findAll('h6')
    for idx, h6 in enumerate(ja_names_h6):
        if idx < len(ja_names_h6) - 1:
            ja_name = h6.text.lower().replace(' ', '-').strip()
            ja_names.append(ja_name)
print(len(ja_names))
```

```python
ja_info_all = []
```

```python
# ja_name_crawed = []
```

```python
url_base = 'https://japanpornjav.com/jav/'
for ja_name in ja_names:
    if ja_name == 'ayuri-sonoda':
        url = url_base + 'ayrui-sonoda'
    elif ja_name == 'chiharu-minagawa':
        url = url_base + 'chiharu-managawa'
    elif ja_name == 'ema-shiiba':
        url = url_base + 'shiiba-ema'
    elif ja_name == 'hikari-aozora':
        url = url_base + 'hikaro-aozora'
    elif ja_name == 'miyu-yanagi':
        url = url_base + 'mayu-yanagi'
    elif ja_name == 'shion-utsunomiya-(rion)':
        url = url_base + 'shion-utsunomiya-rion'
    elif ja_name == 'tina-yuzuki-(rio)':
        url = url_base + 'tina-yuzuki-rio'
    elif ja_name == 'yui-makina':
        url = url_base + 'yui-makino'
    else:
        url = url_base + ja_name
    
    print(url)
    try:
        request_response = requests.get(url).text
        soup_obj = BeautifulSoup(request_response,'lxml')
        # print(request_response)
        ja_info_div = soup_obj.find('div', {'class':'d-flex'})
        # print(ja_info_div)

        name = ja_info_div.h5.text
    except:
        try:
            url = url_base + ja_name.split('-')[1]
            request_response = requests.get(url).text
            soup_obj = BeautifulSoup(request_response,'lxml')
            ja_info_div = soup_obj.find('div', {'class':'d-flex'})

            name = ja_info_div.h5.text
        except:
            url = url_base + ja_name.replace('-','_')
            request_response = requests.get(url).text
            soup_obj = BeautifulSoup(request_response,'lxml')
            ja_info_div = soup_obj.find('div', {'class':'d-flex'})

            name = ja_info_div.h5.text

    img_url = img_url = ja_info_div.img['src']
    ja_info_p = ja_info_div.findAll('p')
    birthday = ja_info_p[0].text
    blood_type = ja_info_p[1].text
    breast_size = ja_info_p[2].text
    waist_size = ja_info_p[3].text
    hip_size = ja_info_p[4].text
    height = ja_info_p[5].text
    number_hd_video_online = ja_info_p[6].text

    ja_info_all.append([name, img_url, birthday, blood_type, breast_size, waist_size, height, number_hd_video_online])

    # download image
    r = requests.get(img_url, allow_redirects=True)
    open('images/'+ ja_name + '.jpg', 'wb').write(r.content)

    # ja_name_crawed.append(ja_name)
    
```

```python
df_ja_info = pd.DataFrame([ja for ja in ja_info_all])
df_ja_info.columns = ['Name', 'Image_URL', 'Birthday', 'Blood_Type', 'Breast_Size', 'Waist_Size', 'Height', 'Number_HD_Video_Online']
print(df_ja_info.shape)
df_ja_info.head()
```

```python
df_ja_info.to_csv('ja_information.csv')
```

# Merge data:
## - Data 1: Download from https://www.kaggle.com/twopothead/japanese-pornstars-and-adult-videos
## - Data 2: Above

```python
df_ja_1 = pd.read_csv('dataset/actress.csv')
df_ja_1.head()
```

```python
df_ja_1.drop(['id', 'imgurl', 'birthday', 'height', 'waist', 'bust'], axis=1, inplace=True)
df_ja_1.head()
```

```python
df_ja_1.rename(columns={'name':'Name', 'cup_size':'Cup_Size', 'hips':'Hips', 'birthplace':'Home_Town', 'hobby':'Hobby'}, inplace=True)
df_ja_1.head()
```

```python
df_ja_2 = pd.read_csv('dataset/ja_information.csv', index_col=[0])
df_ja_2.head()
```

```python
df_ja_2['Name'] = df_ja_2['Name'].apply(lambda x: x.split('/')[1].strip())
df_ja_2.head()
```

```python
df_ja_2['Birthday'] = df_ja_2['Birthday'].apply(lambda x: x.split(':')[1].strip() if pd.isnull(x) == False else x)
df_ja_2['Blood_Type'] = df_ja_2['Blood_Type'].apply(lambda x: x.split(':')[1].strip() if pd.isnull(x) == False else x)
df_ja_2['Breast_Size'] = df_ja_2['Breast_Size'].apply(lambda x: x.split(':')[1][:-3] if pd.isnull(x) == False else x)
df_ja_2['Waist_Size'] = df_ja_2['Waist_Size'].apply(lambda x: x.split(':')[1][:-3] if pd.isnull(x) == False else x)
df_ja_2['Height'] = df_ja_2['Height'].apply(lambda x: x.split(':')[1][:-3] if pd.isnull(x) == False else x)
df_ja_2['Number_HD_Video_Online'] = df_ja_2['Number_HD_Video_Online'].apply(lambda x: x.split(':')[1] if pd.isnull(x) == False else x)
df_ja_2.head()
```

```python
df_ja = df_ja_2.merge(df_ja_1, on='Name', how='left')
df_ja.head()
```

```python
df_ja['Hips'] = df_ja['Hips'].apply(lambda x: x[:-2] if pd.isnull(x) == False else x)
```
